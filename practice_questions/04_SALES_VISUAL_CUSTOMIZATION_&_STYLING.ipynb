```python
import pandas as pd
import matplotlib.pyplot as plt

sales_df = pd.DataFrame({
    "Month": ["Jan","Feb","Mar","Apr","May","Jun","Jul","Aug","Sep","Oct","Nov","Dec"],
    "Revenue": [45000,52000,48000,61000,72000,68000,75000,82000,79000,88000,92000,97000],
    "Orders": [120,140,130,160,180,170,200,220,210,240,260,280],
    "Returns": [5,7,6,8,10,9,11,12,10,13,14,15],
    "Region": ["North","North","South","South","West","West","East","East","North","South","West","East"]
})

sales_df
```

```python
# Plot Revenue vs Month
# Add:
# Title
# X label
# Y label
# Change:
# linestyle → dashed
# marker → circle
# STYLING CONTROL
# Plot Orders trend
# Apply:
# Custom color
# linewidth = 3
# markersize = 8
plt.plot(sales_df["Revenue"],sales_df["Month"],linestyle = "--",marker = "o",color = "grey",linewidth = 2, markersize = 8)
plt.title("Revenue VS Month")
plt.xlabel("Revenue",color = "purple")
plt.ylabel("Month",color = "red")
# Add grid to Revenue graph
plt.grid()

```

```python
high_rev = sales_df["Revenue"].max()
hig_rev_month = sales_df.loc[sales_df["Revenue"].idxmax(),"Month"]
plt.figure(figsize=(10,5))
plt.plot(sales_df["Month"],sales_df["Revenue"],color = "grey",linestyle = ":")
plt.scatter(hig_rev_month,high_rev,s = 150)
plt.title("Monthly Revenue Analysis")
plt.xlabel("Month")
plt.ylabel("Revenue")
plt.grid()
plt.show()
```
