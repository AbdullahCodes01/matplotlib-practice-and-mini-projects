```python
x = [1,2,3,4,5]
y = [100,200,150,300,250]

plt.plot(x,y,color = "red", marker = "o",linestyle = "--")
plt.title("Sales analysis")
plt.xlabel("Days")
plt.ylabel("Sales")
plt.show()
```
