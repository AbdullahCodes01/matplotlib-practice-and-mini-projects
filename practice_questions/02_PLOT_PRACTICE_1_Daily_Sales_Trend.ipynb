```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
sales_trend = pd.DataFrame({
"Day": [1,2,3,4,5],
"Sales": [2000,3500,1000,1600,2400]
})
sales_trend
```

```python
# Make simple line plot
# X = Day
# Y = Sales

plt.plot(sales_trend["Day"],sales_trend["Sales"])
plt.show()
```

```python
# Add:
# âœ” Title â†’ "Weekly Sales Trend"
# âœ” X label â†’ Day
# âœ” Y label â†’ Sales

plt.plot(sales_trend["Day"],sales_trend["Sales"])
plt.title("Weekly Sales Trend")
plt.xlabel("Day")
plt.ylabel("Sales")
plt.show()
```

```python
# Add marker â†’ circle
# Make line dashed
# Change line color
# Add grid
plt.plot(sales_trend["Day"],sales_trend["Sales"],linestyle = "--",color = "pink",marker = "o")
plt.title("Weekly Sales Trend",color = "red")
plt.xlabel("Day")
plt.ylabel("Sales")
plt.grid()
plt.show()
```

```python
# Q7
# Find:
#  Highest sales day
#  Mark that point visually (marker use)
max_sale = sales_trend["Sales"].max()
max_day = sales_trend.loc[sales_trend["Sales"].idxmax(),"Day"]
```

```python
plt.plot(sales_trend["Day"],sales_trend["Sales"],marker = "o",linestyle = "--")
plt.scatter(max_sale,max_day, s = 200)
plt.title("Weekly Sales Analysis",color = "blue")
plt.xlabel("Day")
plt.ylabel("Sales")
plt.show()
```
