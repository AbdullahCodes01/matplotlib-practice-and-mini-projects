```python
import pandas as pd
import matplotlib.pyplot as plt 
```

```python
revenue_df = pd.DataFrame({
    "Month": ["Jan","Feb","Mar","Apr","May","Jun"],
    "Revenue": [50000, 52000, 48000, 60000, 65000, 70000]
})

revenue_df
```

```python
# Plot Revenue Trend
# find
# Revenue growth trend visually
# Highlight highest revenue month
max_revenue = revenue_df["Revenue"].max()
max_month = revenue_df.loc[revenue_df["Revenue"].idxmax(),"Month"]
plt.plot(revenue_df["Month"],revenue_df["Revenue"],linestyle = "--",color = "brown")
plt.scatter(max_month, max_revenue,s = 150)
plt.title("Revenue Growth Trend")
plt.xlabel("Month")
plt.grid()
plt.ylabel("Revenue")
plt.show()
```
