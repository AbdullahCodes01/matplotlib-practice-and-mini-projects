```python
import matplotlib.pyplot as plt

months = ["Jan","Feb","Mar","Apr","May"]
sales = [50,60,55,70,90]
```

```python
plt.plot(months,sales,color = "purple", linestyle = ":",linewidth = 2,marker ="o")
plt.title("Monthly Sales Trend",color = "blue")
plt.xlabel("Month",color = "red")
plt.ylabel("Sales",color = "green")
plt.grid() #Grid (Professional Look)
plt.figure(figsize=(8,5))         #Figure Size (Dashboard Level)
plt.show()
```

```python
# Legend (Multi Line Graph)

sales2 = [60,40,70,80,50]
plt.figure(figsize=(8,5))  #Figure Size (Dashboard Level)
plt.plot(months,sales,label = "2024 sales" ,linestyle = ":",linewidth = 2,marker ="o")
plt.plot(months,sales2,label= "2025 sales" ,linestyle = "--",marker = "o")
plt.title("Monthly Sales Trend",color = "blue")
plt.xlabel("Month",color = "red")
plt.ylabel("Sales",color = "green")
# plt.xlim("auto")
plt.ylim()
plt.grid() #Grid (Professional Look)
plt.legend()

plt.show()
```
