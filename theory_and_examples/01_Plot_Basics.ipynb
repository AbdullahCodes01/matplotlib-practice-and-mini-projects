```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# (Line Plot)
# plt.plot(x, y)
# x-axis → x list
# y-axis → y list
x = [1,2,3,4]
y = [10,20,30,40]
plt.plot(x,y)
# Title & Labels
plt.title("Sales Growth")
plt.xlabel("Days")
plt.ylabel("Sales")
plt.show()
```

```python
# Marker Add (Data Points Show)
plt.plot(x,y, marker = "o")
plt.show()
```

```python
plt.plot(x,y,linestyle = "--") # Line Style Change
plt.show()
```

```python
plt.plot(x,y,color = "red",linestyle = "--",marker= "o") #color change
plt.show()
```

```python
plt.plot(x,y,color = "green",linestyle ="--",marker = "o")
plt.title("Sales Growth")
plt.xlabel("Days")
plt.ylabel('Sale')
plt.show()
```
